```python
import whisper

# Charger le modèle : "small" = rapide, "medium" = plus précis
model = whisper.load_model("small")

# Transcrire ton audio (ici en français)
result = model.transcribe("/content/reunion.mp3", language="fr")

# Sauvegarder le texte dans un fichier
with open("transcription.txt", "w", encoding="utf-8") as f:
    f.write(result["text"])

print("✅ Transcription terminée ! Aperçu :\n")
print(result["text"][:800])  # affiche les 800 premiers caractères
```
